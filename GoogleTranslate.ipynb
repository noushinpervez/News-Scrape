```python
import pandas as pd
from easygoogletranslate import EasyGoogleTranslate
```

```python
e_df = pd.read_csv('/content/drive/My Drive/summarization_english.csv')

e_df = e_df.drop(['URL', 'Headline', 'Article'], axis=1)

e_df.head()
```

```python
total_rows = len(e_df)
chunk_size = 5000

total_rows
```

```python
translator = EasyGoogleTranslate(
    source_language='en',
    target_language='bn',
    timeout=10
)
```

```python
for i in range(0, total_rows, chunk_size):
    chunk_df = e_df.iloc[i:i+chunk_size]
    translated_summaries = []

    for idx, row in chunk_df.iterrows():
        translated_summary = translator.translate(row['Summary'])

        if idx % 1000 == 0:
            print('Row', idx, 'Original:', row['Summary'])
            print('Row', idx, 'Translated:', translated_summary)
            print()

        translated_summaries.append({'english': row['Summary'], 'bangla': translated_summary})

    translated_df = pd.DataFrame(translated_summaries)
    file_path = f'/content/drive/My Drive/google_translation_{i//chunk_size + 1}.csv'
    translated_df.to_csv(file_path, encoding='utf-8', index=False)
    print(f'DataFrame saved to: {file_path}')
```
